**Concept of Function**
- A function is a block of code grouped into a single unit which only runs when it is called.
- A function can be called many times independently.
- The purpose of the function is to break a large and complex program into small parts with their respective tasks.

**Arguments vs Parameters**
- Function can accept data/informations called arguments or parameters.
- The terms parameter and argument can be used for the same thing: Information that are passed into a function.
- From a function's perspective in programming:
  - A parameter is the variable listed inside the parentheses in the function definition.
  - An argument is the value that is sent to the function when it is called.

**Types of Functions in Python**
- Empty Function
- Function with Parameters (Required and Optional)
- Function with Arbitrary Arguments
- Function with Keyword Arguments
- Function with Arbitrary Keyword Arguments
- Function that Return a Value (Yes or No)
- Recursion Function (A defined function can call itself)

**Non-existent Types of Functions in Python**
- Overloading Function (A function with the same name but with different parameters)
- Overriding Function (A function with the same name but in a different class (inherited class))

####**Function with Parameters:**

````python
#function with required parameter
def welcome(name):
  print(f'Hi {name}, welcome!')

welcome(1)
welcome(2.0)
welcome(True)
welcome('Alpha')
````

````python
#function with required parameters
def welcome(name, place):
  print(f'Hi {name}, welcome to {place}')


welcome('Alpha', 'Python')
welcome('Beta', 'Visual Studio Code')
welcome('Gamma', 'Jupyter Notebook')
````

````python
#function with optional parameter
def welcome(name, place='Google Colab'):
  print(f'Hi {name}, welcome to {place}')

welcome('Alpha')
welcome('Beta', 'Google Docs')
````

````python
#function with optional parameters
def welcome(name, place='Google Colab', time=0.0):
  print('Hi {0}, welcome to {1} at {2:.2f}'.format(name, place, time))

welcome('Alpha')
welcome('Beta', 'Google Docs')
welcome('Gamma', 'Google Drive', 12.00)
````

####**Function with Arguments:**

````python
#function with arbitrary arguments (if you do not know how many arguments)
#arbitrary arguments -> *args 
def introduce(*species):
  for i in species:
    print(f'I am {i}')

introduce("Human", "Robot", "Alien")
````

````python
#function with keyword arguments (if you want to ignore the order of arguments)
#keyword arguments -> *kwrgs 

def introduce(species3, species2, species1):
  print(f'You are {species1}')
  print(f'You are {species2}')
  print(f'You are {species3}')

introduce(species1 = "Human", species2 = "Robot", species3 = "Alien")
````

````python
#function with arbitrary keyword Arguments (if you do not know how many arguments and you want to ignore the order of arguments)

def introduce(**species):
    print('We are ' + species['species1'])
    print('We are ' + species['species2'])
    print('We are ' + species['species3'])

introduce(species3='Alien', species2='Robot', species1='Human')
````

#### **Function with Return Value:**

````python
#Function that return a value
def add2values(a, b):
  result = a + b
  return result 

#case01 (just call the function)
add2values(1, 2)

#case02 (store the function in a variable)
a = add2values(1, 2)

#case3 (print the function)
print("1 + 2 =", add2values(1, 2)) 
````

````python
#Function that return no value
def add2values(a, b):
  result = a + b
  print(a, '+', b, '=', result) 

add2values(1, 2)
````

####**Recursion Function:**

````python
#recursion function with limit 
def show_intervals(start, end):
  print(start)
  start = start + 1

  if(start <= end):
     show_intervals(start, end)

show_intervals(1, 10)
````

````python
#recursion function with no limit
def show_interval(value):
   print(value)   
   value = value + 1
   show_interval(value)

show_interval(2)
````
