```python
# random txt generator
import random
import string

def generate_random_string(length):
    # Generate a random string of specified length
    letters = string.ascii_letters + string.digits + string.punctuation
    return ''.join(random.choice(letters) for i in range(length))

def create_random_text_file(file_name, file_size):
    # Create a text file with random characters of specified size
    with open(file_name, 'w') as f:
        f.write(generate_random_string(file_size))

# Example usage
create_random_text_file('text.txt', 64) # Creates a random text file of 64 characters
```

```python
# 4. Enumerate function
names = ['Al', 'El', 'Dul']

for index, name in enumerate(names, start=1):
    print(index, name)
```

```python
# 5. Zip function
superhero_nicknames = ['Superman', 'Spiderman', 'Batman', 
                   'Wonder Woman', 'Captain America']
superhero_realnames = ['Clark Kent', 'Peter Parker', 'Bruce Wayne', 
                        'Diana Prince', 'Steve Rogers']

for nickname, realname in zip(superhero_nicknames, superhero_realnames):
    print(f'{nickname} is {realname}')
```

```python
# 6. Unpacking values
a, _ = (1, 2) # _ is a var that we arn't going to use 
print(a)

a, b, *_ = (1, 2, 3, 4, 5) # *_ save the rest of values 
print(a, b, _)

a, b, *_, d = (1, 2, 3, 4, 5) # *_ save the rest of values except the last value 
print(a, b, _, d)
```

```python
# 7. Getting and setting attributes
class Person():
    pass

person = Person()

# set key and value for an object directly
setattr(person, 'first_name', 'Muhammad')

# get key and value for an object directly
print(getattr(person, 'first_name'))

print('='*30)

# set dict as attributes to the object
person_info = {'first_name': 'Muhammad', 'last_name': 'Haerul'}

for key, value in person_info.items():
    setattr(person, key, value)
    
for key in person_info.keys():
    print(getattr(person, key))
```
